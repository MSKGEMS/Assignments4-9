```python
"""How-to-count-distance-to-the-previous-zero"""
#Assignment7 Q1
import numpy as np
import pandas as pd

df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
izero = np.r_[-1, (df['X'] == 0).nonzero()[0]] # indices of zeros
idx = np.arange(len(df))
df['Y'] = idx - izero[np.searchsorted(izero - 1, idx) - 1]
print(df)
```

```python
 """Create a DatetimeIndex that contains each business day of 2015 and use it to index a
Series of random numbers."""
#Assignment7 Q2

import numpy as np
import pandas as pd
dti = pd.date_range(start='2015-01-01', end='2015-12-31', freq='D') 
s = pd.Series(np.random.rand(len(dti)), index=dti)

dti,s
```

```python
"""Find the sum of the values in s for every Wednesday"""
#Assignment7 Q3

s[dti.weekday == 2].sum() 
```

```python
"""Average For each calendar month"""
#Assignment7Q4

s.resample('M').mean()
```

```python
"""For each group of four consecutive calendar months in s, find the date on which the highest
value occurred."""
#Assignment7 Q5

s.groupby(pd.Grouper(freq = '4M')).idxmax()
```

```python
#Read the dataset from the below link
import pandas as pd
#Baby names data from the link 
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
df.head()
df

#droping the unamed column
df = df.drop('Unnamed: 0', 1)
print(df)

```

```python
#3) Show the top 5 most preferred names
# you don't want to sum the Year column, so you delete it
del df['Year']

# group the data
names = df.groupby("Name").sum()

# sort it from the biggest value to the smallest one
names.sort_values("Count", ascending = 0).head()

```

```python
#4) What is the median name occurrence in the dataset
names[names.Count == names.Count.median()]

```
